```python
import numpy as np
import pandas as pd

import kraft
```

## cluster

```python
point_x_dimension = np.asarray(
    (
        (
            9,
            19,
            9,
            19,
            9,
        ),
        (
            0,
            10,
            0,
            10,
            0,
        ),
        (
            1,
            11,
            1,
            11,
            1,
        ),
        (
            8,
            18,
            8,
            18,
            8,
        ),
    )
)

print(point_x_dimension)

for n_cluster in (
    None,
    3,
):

    print("=" * 80)

    print(n_cluster)

    print(
        kraft.clustering.cluster(
            point_x_dimension,
            n_cluster=n_cluster,
        )
    )
```

## get_coclustering_distance

```python
point_x_clustering = np.array(
    (
        (
            1,
            1,
            np.nan,
            np.nan,
        ),
        (
            0,
            0,
            1,
            1,
        ),
        (
            1,
            0,
            np.nan,
            1,
        ),
    )
).T

print(point_x_clustering)

for min_n_clustered in (
    0,
    1,
):

    print("=" * 80)

    print(min_n_clustered)

    try:

        print(
            kraft.clustering.get_coclustering_distance(
                point_x_clustering,
                min_n_clustered,
            )
        )

    except AssertionError as error:

        print(error)
```

## cluster_hierarchical_clusterings

```python
point_x_dimension = np.random.random_sample(
    size=(
        80,
        160,
    )
)

point_x_dimension[
    0::4,
    :,
] *= 1

point_x_dimension[
    1::4,
    :,
] *= 2

point_x_dimension[
    2::4,
    :,
] *= 3

point_x_dimension[
    3::4,
    :,
] *= 4


point_x_dimension = pd.DataFrame(
    point_x_dimension,
    index=pd.Index(
        ("Label {}".format(i) for i in range(point_x_dimension.shape[0])),
        name="Axis 0",
    ),
    columns=pd.Index(
        ("Lbael {}".format(i) for i in range(point_x_dimension.shape[1])),
        name="Axis 1",
    ),
)

for n_cluster in (
    2,
    3,
    4,
):

    for optimal_ordering in (
        False,
        True,
    ):

        print("=" * 80)

        print(
            n_cluster,
            optimal_ordering,
        )

        kraft.clustering.cluster_hierarchical_clusterings(
            point_x_dimension,
            n_cluster,
            optimal_ordering=optimal_ordering,
        )
```
