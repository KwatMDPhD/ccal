```python
import kraft
import numpy as np
```

```python
vector_with_nan_0 = np.asarray(
    (
        np.nan,
        0,
        1,
        2,
    )
)

vector_with_nan_1 = np.asarray(
    (
        4,
        8,
        0,
        np.nan,
    )
)
```

## get_entropy

```python
try:

    kraft.information.get_entropy(vector_with_nan_0)

except AssertionError as error:

    print(error)
```

```python
xs = []

ys = []

for power in range(10):

    n = 2 ** power

    xs.append(n)

    vector = np.random.random_sample(size=n)

    ys.append(kraft.information.get_entropy(vector))

kraft.plot.plot_x_y(
    xs,
    ys,
    "N",
    "Entropy",
)
```

```python
xs = []

ys = []

for power in range(10):

    n = 2 ** power

    xs.append(n)

    vector = np.full(
        n,
        1,
    )

    ys.append(kraft.information.get_entropy(vector))

kraft.plot.plot_x_y(
    xs,
    ys,
    "N",
    "Entropy",
)
```

## get_kld

```python
try:

    kraft.information.get_kld(
        vector_with_nan_0,
        vector_with_nan_1,
    )

except AssertionError as error:

    print(error)
```

```python
xs = []

ys = []

for power in range(10):

    n = 2 ** power

    xs.append(n)

    vector_0 = np.random.random_sample(size=n)

    vector_1 = np.random.random_sample(size=n)

    klds = kraft.information.get_kld(
        vector_0,
        vector_1,
    )

    ys.append(klds.sum())

kraft.plot.plot_x_y(
    xs,
    ys,
    "N",
    "KLD",
)
```
