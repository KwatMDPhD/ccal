```python
import os
import sys
import importlib
```

```python
REPOSITORIES_DP = '../..'
```

```python
for i, fn in enumerate(sorted(os.listdir(REPOSITORIES_DP))):
    print('\n{} {}'.format(i, fn))

    if fn in [
            '.DS_Store',
            'environment',
            'gpg',
            'jupyter_gui',
            'kwatme.com',
    ]:
        continue

    sys.path.insert(0, os.path.join(REPOSITORIES_DP, fn))

    for mn in os.listdir(os.path.join(REPOSITORIES_DP, fn, fn)):

        if mn.endswith('.py'):

            command = 'from {} import {}'.format(fn, mn[:-3])
            print('\t' + command)
            exec(command)

    sys.path.pop(0)
```
